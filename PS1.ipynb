```python
# Chap.2 1.29
import numpy as np
import matplotlib.pyplot as plt
u_0=np.matrix('1; 0')
u=u_0
A=np.matrix('0.8 0.3; 0.2 0.7')

array = np.transpose(u_0)

for i in range(50):
  u = A*u
  # print(f'u_{i}: {u[0]}')
  transposed_u = np.transpose(u)
  array = np.vstack([array,transposed_u])
  i += 1

plt.plot(array[:,0])
plt.plot(array[:,1])
```
